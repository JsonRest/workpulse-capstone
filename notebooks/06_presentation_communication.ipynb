## 1. 📊 Technical Presentation

### Purpose
Communicate the **methodology, model architecture, evaluation, and technical decisions** to a peer audience who understands ML concepts.

### Format Options
Three formats are provided for flexibility:

| Format | File | Best For |

### Slides (11 slides)

| # | Slide Title | Content |
|---|------------|----------|
| 1 | Title | Project name, subtitle, programme |
