## 5. 🧪 CI/CD Pipeline

The repository includes `.github/workflows/ci.yml` that runs automatically on every push and pull request:

```yaml
name: WorkPulse CI
on: [push, pull_request]
jobs:
  lint-and-test:
    runs-on: ubuntu-latest
    steps:
      - uses: actions/checkout@v4
      - uses: actions/setup-python@v5
        with: {python-version: '3.9'}
      - run: pip install flake8 pytest numpy pandas scikit-learn xgboost
      - run: flake8 src/ tests/ --max-line-length=120
      - run: pytest tests/ -v --tb=short
```

### Tests (11 total)

| Test | What It Checks |
|------|---------------|
| `test_shape` | Dataset has correct row count and column count |
| `test_binary_target` | Target is binary (0 or 1) |
| `test_target_balance` | Positive class is ~35% (within 30-40%) |
| `test_no_nulls` | No missing values |
| `test_feature_ranges` | Features within expected bounds |
| `test_reproducibility` | Same seed produces identical data |
| `test_different_seeds` | Different seeds produce different data |
| `test_split_sizes` | 80/20 train/test split |
| `test_stratified` | Split maintains target distribution |
| `test_scaler` | Scaled training data has mean≈0, std≈1 |
| `test_feature_count` | Correct number of features (13) |
