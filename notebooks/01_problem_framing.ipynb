## 1. 📌 Executive Summary

Employee burnout is one of the most costly and under-detected challenges facing organisations today. According to Gallup's 2024 State of the Global Workplace report, **approximately 23% of employees report feeling burned out at work very often or always**, while a further 44% report feeling burned out sometimes — representing a silent productivity crisis affecting nearly two-thirds of the global workforce.

**WorkPulse** is an AI-powered Early Warning System that uses machine learning to identify employees at high risk of burnout *before* it escalates into attrition, medical leave, or performance collapse. By surfacing risk signals early, HR teams and managers can intervene with targeted support — reducing the human and financial cost of burnout at scale.

This capstone project frames burnout risk prediction as a **supervised binary classification problem**, combining multiple public HR datasets with synthetic data augmentation to build a robust, fair, and explainable predictive system.

## 4. 🤖 ML Task Definition

### 4.1 Task Type: Supervised Binary Classification

This is a **supervised binary classification** problem.

**Justification:**
- We have labelled training data (burnout scores and attrition labels from public HR datasets)
- The output is a **discrete binary decision**: High Risk vs. Low Risk
- The goal is to learn a mapping function `f(X) → y` where `y ∈ {0, 1}`
- A continuous risk probability score `P(y=1 | X)` will also be produced for ranking and thresholding

**Why not regression?**  
While burnout has a continuous underlying score, the business action is binary — an HR manager either initiates an intervention or doesn't. A classification output with an accompanying probability score gives both precision and actionability.

**Why not clustering (unsupervised)?**  
We have access to labelled outcome data (burnout scores, attrition flags). Supervised learning will yield a more reliable, directly actionable predictive system than unsupervised grouping alone. Clustering will be used as a *supplementary* technique during EDA to discover natural employee segments.

## 5. 📊 Success Metrics

### 5.1 Why Metric Selection Matters Here

The burnout dataset will have **class imbalance** — not every employee is burned out. This means **accuracy alone is misleading**: a model that predicts "Low Risk" for everyone could achieve 77% accuracy while being completely useless.

We prioritise metrics that are robust to imbalance and aligned with the cost of each error type:

| Error Type | In Our Context | Business Cost |
|-----------|---------------|---------------|
| **False Negative** (missed high-risk employee) | Employee burns out, leaves, or gets sick | HIGH — turnover cost, lost productivity, human suffering |
| **False Positive** (flagged as at-risk but isn't) | Unnecessary check-in from manager | LOW — minor manager time investment |

→ **We prioritise Recall** (catching true positives) over Precision. Missing a burned-out employee is far costlier than a false alarm.

### 5.2 Primary & Secondary Technical Metrics

| Metric | Target | Why |
|--------|--------|-----|
| **AUC-ROC** *(Primary)* | ≥ 0.85 | Best overall measure of classifier discrimination; robust to imbalance |
| **Recall (Sensitivity)** *(Primary)* | ≥ 0.80 | Minimise missed high-risk employees |
| **F1-Score** | ≥ 0.78 | Balanced measure of precision and recall |
| **Precision** | ≥ 0.75 | Ensure flagged employees have real cause for concern |
| **Average Precision (PR-AUC)** | ≥ 0.80 | Especially informative under class imbalance |
| **Log Loss** | ≤ 0.35 | Measures calibration of probability outputs |

### 5.3 Model Comparison Framework

All models will be evaluated using the same framework:
- **Stratified 5-Fold Cross-Validation** on the training set
- **Hold-out Test Set** (20% of data, never seen during training or tuning)
- **McNemar's Test** for statistical significance between top-performing models

## 7. ⚠️ Assumptions & Constraints

Documenting assumptions is a mark of professional, honest data science. Every model rests on assumptions — being explicit about them prevents overconfident conclusions.

### 7.1 Data Assumptions

| Assumption | Implication if Wrong |
|------------|---------------------|
| Public HR datasets are representative of real workplace dynamics | Model may not generalise to all industries or geographies |
| Synthetic augmentation (CTGAN) preserves real statistical relationships | Inflated performance metrics if synthetic data is too "perfect" |
| Burnout score from Employee Burnout dataset is a valid proxy for true burnout | Target variable may contain label noise |
| Features are measured at a single point in time (cross-sectional) | Cannot capture temporal burnout trajectory without time-series data |

### 7.2 Modelling Constraints

| Constraint | Mitigation Strategy |
|-----------|--------------------|
| **Class Imbalance**: ~23% positive class | SMOTE oversampling + class_weight='balanced' + AUC as primary metric |
| **No longitudinal data**: single snapshot per employee | Frame as point-in-time risk assessment; note limitation clearly |
| **Sensitive attributes present**: gender, age, department | Full fairness audit in Step 5; propose mitigation strategies |
| **Synthetic majority of data**: ~460K of 500K rows are synthetic | Validate model separately on real data only; report both performance figures |

### 7.3 Ethical Constraints

> ⚠️ **Important Ethical Note:** An employee burnout prediction system has significant ethical implications. The following constraints will govern this project:

1. **Transparency**: Employees should be informed if such a system is in use
2. **No punitive use**: Risk scores must not be used for performance management or termination decisions
3. **Human-in-the-loop**: All model outputs should be reviewed by trained HR professionals before action
4. **Data minimisation**: Only features necessary for prediction should be collected
5. **Fairness obligation**: The model must not systematically disadvantage protected groups

````python
# ============================================================
# BUSINESS ROI CALCULATOR
# ============================================================
# Translate model performance into financial impact.
# Adjust assumptions to match your target organisation.

def calculate_roi(
    org_size=5000,
    burnout_rate=0.23,
    avg_salary=60_000,
    replacement_cost_pct=1.0,
    model_recall=0.80,
    intervention_success_rate=0.40
):
    """
    Calculate estimated annual cost savings from WorkPulse deployment.

    Parameters:
    -----------
    org_size                  : int   — Total number of employees
    burnout_rate              : float — % of employees at risk annually
    avg_salary                : float — Average annual salary (USD)
    replacement_cost_pct      : float — Replacement cost as % of salary (1.0 = 100%)
    model_recall              : float — Model recall (% of at-risk employees caught)
    intervention_success_rate : float — % of flagged employees where intervention prevents attrition

    Returns:
    --------
    dict: ROI breakdown
    """
    at_risk_employees     = org_size * burnout_rate
    employees_caught      = at_risk_employees * model_recall
    attrition_prevented   = employees_caught * intervention_success_rate
    replacement_cost      = avg_salary * replacement_cost_pct
    annual_savings        = attrition_prevented * replacement_cost

    return {
        'Organisation Size'         : f"{org_size:,} employees",
        'At-Risk Employees/Year'    : f"{at_risk_employees:,.0f} ({burnout_rate*100:.0f}%)",
        'Caught by Model'           : f"{employees_caught:,.0f} ({model_recall*100:.0f}% recall)",
        'Attrition Prevented'       : f"{attrition_prevented:,.0f}",
        'Cost per Replacement'      : f"${replacement_cost:,.0f}",
        'Estimated Annual Savings'  : f"${annual_savings:,.0f}",
        'Payback Period'            : "< 1 month (vs ~$200K build cost)",
    }


# Run the ROI calculation
roi = calculate_roi()

print("=" * 55)
print("   WORKPULSE BUSINESS ROI ESTIMATE")
print("=" * 55)
for key, value in roi.items():
    print(f"  {key:<30} : {value}")
print("=" * 55)

# Sensitivity analysis — conservative vs optimistic
print("\n  SENSITIVITY ANALYSIS")
print("-" * 55)
for label, recall, success in [
    ("Conservative (Recall=0.70, Success=30%)", 0.70, 0.30),
    ("Base Case    (Recall=0.80, Success=40%)", 0.80, 0.40),
    ("Optimistic   (Recall=0.90, Success=50%)", 0.90, 0.50),
]:
    r = calculate_roi(model_recall=recall, intervention_success_rate=success)
    print(f"  {label}")
    print(f"    → Savings: {r['Estimated Annual Savings']}")
````
