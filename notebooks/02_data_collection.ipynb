```python
# ── 3.3 Missing Value Analysis ────────────────────────────────
def missing_summary(df, name='Dataset'):
    """Return a clean missing value summary DataFrame."""
    miss_count = df.isnull().sum()
    miss_pct   = 100 * miss_count / len(df)
    result = pd.DataFrame({
        'Missing Count' : miss_count,
        'Missing %'     : miss_pct.round(2),
        'Dtype'         : df.dtypes
    })
    result = result[result['Missing Count'] > 0].sort_values('Missing %', ascending=False)
    if result.empty:
        print(f'  {name}: No missing values found ✅')
    else:
        print(f'  {name}: {len(result)} columns with missing values')
        display(result)
    return result

print('IBM HR — Missing Value Summary')
ibm_missing = missing_summary(ibm_raw, 'IBM HR')
```

```python
# ── 3.4 Target Variable Distribution ─────────────────────────
attrition_counts = ibm_raw['Attrition'].value_counts()
attrition_pct    = ibm_raw['Attrition'].value_counts(normalize=True) * 100

print('IBM HR — Target Variable: Attrition')
print('=' * 40)
for label in ['Yes', 'No']:
    print(f'  {label:5s} : {attrition_counts[label]:>5,}  ({attrition_pct[label]:.1f}%)')
print(f'  Class imbalance ratio: 1 : {attrition_counts["No"]/attrition_counts["Yes"]:.1f}')

# Quick visualisation
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Bar chart
colors = ['#e74c3c', '#2ecc71']
attrition_counts.plot(kind='bar', ax=axes[0], color=colors, edgecolor='white', width=0.5)
axes[0].set_title('IBM HR: Attrition Distribution', fontweight='bold', pad=12)
axes[0].set_xlabel('Attrition', labelpad=8)
axes[0].set_ylabel('Employee Count')
axes[0].tick_params(axis='x', rotation=0)
for i, v in enumerate(attrition_counts):
    axes[0].text(i, v + 10, f'{v:,}\n({attrition_pct.iloc[i]:.1f}%)',
                 ha='center', fontsize=10, fontweight='bold')

# Pie chart
axes[1].pie(attrition_counts, labels=attrition_counts.index,
            colors=colors, autopct='%1.1f%%', startangle=90,
            wedgeprops={'edgecolor': 'white', 'linewidth': 2})
axes[1].set_title('IBM HR: Attrition Proportion', fontweight='bold', pad=12)

plt.tight_layout()
plt.savefig('data/raw/fig_ibm_target_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print('\n💡 Insight: 16% positive class — significant imbalance requiring SMOTE or class_weight balancing.')
```

## 8. 🧬 Synthetic Augmentation Plan

### 8.1 Why Synthetic Augmentation?

Our three source datasets total ~44,220 rows. While sufficient for a basic model, we target **500,000+ rows** to:
1. Stress-test model robustness at production scale
2. Enable more stable cross-validation across rare subgroups
3. Demonstrate knowledge of data augmentation techniques — a valued industry skill
4. Bridge the gap toward real enterprise HR datasets (which are never public)

### 8.2 Augmentation Method: CTGAN

We will use **CTGAN (Conditional Tabular GAN)** from the `sdv` library — the leading framework for synthetic tabular data generation. CTGAN:
- Handles mixed data types (numeric + categorical) natively
- Conditions on class labels to preserve burnout risk proportions
- Maintains realistic statistical correlations between features
- Is widely used in industry for privacy-preserving data synthesis

```python
# ── 8.3 Augmentation Roadmap ──────────────────────────────────
# This cell documents the plan; full execution happens in Step 3.

AUGMENTATION_PLAN = {
    'method'            : 'CTGAN (Conditional Tabular GAN)',
    'library'           : 'sdv (Synthetic Data Vault)',
    'install'           : 'pip install sdv',
    'input_rows'        : 44_220,
    'target_rows'       : 500_000,
    'synthetic_rows'    : 500_000 - 44_220,
    'class_condition'   : True,   # Condition on burnout_risk to preserve 23% positive rate
    'epochs'            : 300,
    'validation'        : 'Statistical similarity test (KS-test per feature)',
    'privacy_note'      : 'Synthetic data cannot be reverse-engineered to real individuals',
}

print('SYNTHETIC AUGMENTATION PLAN')
print('=' * 50)
for k, v in AUGMENTATION_PLAN.items():
    print(f'  {k:<22} : {v}')

print()
print('SCALE-UP SUMMARY')
print('=' * 50)
print(f'  Source rows      : {AUGMENTATION_PLAN["input_rows"]:>8,}')
print(f'  Synthetic rows   : {AUGMENTATION_PLAN["synthetic_rows"]:>8,}')
print(f'  Total rows       : {AUGMENTATION_PLAN["target_rows"]:>8,}')
print(f'  Scale factor     : {AUGMENTATION_PLAN["target_rows"]/AUGMENTATION_PLAN["input_rows"]:>7.1f}×')
```

```python
# ── 8.4 Augmentation Code Preview ────────────────────────────
# Shown here for planning; executed in Step 3 after schema unification.

augmentation_preview = '''
# ============================================================
# STEP 3 PREVIEW: CTGAN Synthetic Data Generation
# Will be executed after schema unification in Step 3
# ============================================================

from sdv.single_table import CTGANSynthesizer
from sdv.metadata    import SingleTableMetadata

# 1. Define metadata for the unified schema
metadata = SingleTableMetadata()
metadata.detect_from_dataframe(df_unified)

# Mark sensitive columns so CTGAN handles them correctly
metadata.update_column('burnout_risk', sdtype='boolean')
metadata.update_column('gender',       sdtype='categorical')

# 2. Train the synthesizer
synthesizer = CTGANSynthesizer(
    metadata,
    epochs=300,
    verbose=True
)
synthesizer.fit(df_unified)

# 3. Generate synthetic rows
n_synthetic = 500_000 - len(df_unified)
df_synthetic = synthesizer.sample(num_rows=n_synthetic)

# 4. Combine real + synthetic
df_full = pd.concat(
    [df_unified.assign(source='real'), 
     df_synthetic.assign(source='synthetic')],
    ignore_index=True
)

# 5. Validate: KS-test per numeric feature
from scipy.stats import ks_2samp
for col in df_unified.select_dtypes(include=np.number).columns:
    stat, p = ks_2samp(df_unified[col].dropna(), df_synthetic[col].dropna())
    print(f"{col:<35} KS={stat:.3f}  p={p:.3f}  {'✅ OK' if p > 0.05 else '⚠️ Review'}")
'''

print('CTGAN Augmentation Code (to be run in Step 3):')
print(augmentation_preview)
```
