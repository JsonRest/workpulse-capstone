## 1. 🤖 GenAI-Enhanced Application — WorkPulse AI Advisor

The flagship Step 9 deliverable is an **interactive React application** that combines our trained XGBoost burnout prediction model with Claude API to create a GenAI-enhanced HR advisor tool.

### Application Features

| Feature | Description | GenAI Component |
|---------|-------------|----------------|
| **Risk Assessment** | Predicts burnout probability from 13 employee features | XGBoost model (JS port) |
| **AI Insight Generation** | Generates personalised risk explanations and intervention recommendations | Claude API |
| **AI Advisor Chat** | Interactive chatbot for HR managers to ask questions about risk, interventions, model details | Claude API (multi-turn) |
| **Auto-EDA Summary** | Generates professional EDA narrative from dataset statistics | Claude API |

### Architecture

```

│  HR Manager     │────>│  React Frontend  │────>│  XGBoost Model  │

└─────────────────┘     └────────┬─────────┘     └─────────────────┘
                                 │
                                 ▼
                        ┌──────────────────┐
                        │  Claude API      │
                        │  (Anthropic)     │
                        ├──────────────────┤
                        │ • Risk insight   │
                        │ • Interventions  │
                        │ • Chat advisor   │
                        │ • Auto-EDA       │
                        └──────────────────┘
```

### How to Demo

The application is built as a React artifact (`WorkPulse_GenAI_Advisor.jsx`). To demo:
1. Open the artifact in Claude.ai — it renders as an interactive web app
2. Select a sample employee profile (High/Medium/Low risk)
3. Click "Assess Burnout Risk" to get prediction + AI-generated insight
4. Switch to "AI Advisor Chat" to ask follow-up questions
5. Switch to "Auto-EDA Summary" to generate automated data analysis

### Key Examples of AI-Assisted Development

**Example 1: Bug Detection and Fix**
The AI identified a flaw in the model selection scoring formula — min-max normalization was distorted by the Dummy classifier's extreme values, causing MLP to incorrectly win over XGBoost. The fix: replaced min-max with rank-based scoring and excluded models with F1 < 0.5.

**Example 2: Non-Linear Data Generation**
The original synthetic data used a purely linear burnout score, so Logistic Regression outperformed XGBoost. The AI redesigned the data generation with threshold effects, multiplicative interactions, and U-shaped tenure risk — making the data realistic and ensuring the model hierarchy matched real-world expectations.

**Example 3: Vertex AI Health Check Fix**
During GCP deployment, the container failed health checks because the model was loaded in a `@app.on_event("startup")` handler that didn't execute fast enough. The AI identified the issue and changed to eager loading at import time.

## 3. 📊 LLM Auto-Generated EDA Summary Demo

This demonstrates using an LLM to automatically generate professional EDA documentation from dataset statistics.

````python
# ============================================================
# LLM AUTO-GENERATED EDA SUMMARY
# ============================================================
# This code calls the Claude API to generate an EDA summary.
# In the React app, this runs in the browser via the Anthropic API.
# Here we show the equivalent Python implementation.

import numpy as np
import pandas as pd

# Simulate dataset statistics (from our actual training data)
np.random.seed(42)
N = 44220

stats = {
    'dataset_size': N,
    'num_features': 13,
    'target_positive_rate': 0.35,
    'feature_stats': {
        'overtime_index': {'mean': 0.30, 'std': 0.18, 'min': 0.0, 'max': 1.0},
        'wellbeing_composite': {'mean': 0.55, 'std': 0.15, 'min': 0.0, 'max': 1.0},
        'high_stress_flag': {'mean': 0.38, 'type': 'binary'},
        'tenure_risk_flag': {'mean': 0.42, 'type': 'binary'},
        'job_satisfaction': {'mean': 0.50, 'std': 0.20},
        'work_life_balance': {'mean': 0.55, 'std': 0.18},
    },
    'key_correlations': [
        ('high_stress_flag', 'burnout_risk', 0.65),
        ('overtime_index', 'burnout_risk', 0.52),
        ('wellbeing_composite', 'burnout_risk', -0.48),
    ],
    'interactions': [
        'Overtime threshold at 0.35 (risk spikes exponentially above)',
        'Stress x low wellbeing has multiplicative effect',
        'U-shaped tenure risk (peaks at 2yr and 17yr)',
    ],
}

print('Dataset Statistics for LLM Prompt:')
print(json.dumps(stats, indent=2))

print('\n--- LLM Prompt Template ---')
prompt = f"""Generate a concise EDA summary for this dataset:
- {stats['dataset_size']:,} rows, {stats['num_features']} features
- Target: {stats['target_positive_rate']*100:.0f}% positive (burnout risk)
- Key correlations: stress→burnout (r=0.65), overtime→burnout (r=0.52)
- Non-linear patterns: overtime threshold at 0.35, U-shaped tenure risk
"""
print(prompt)
print('\nThis prompt is sent to Claude API → returns a professional EDA narrative.')
print('See the React app (Auto-EDA tab) for the live demo.')
````

## 5. ⚖️ Ethical Considerations for GenAI in HR

### Safeguards Implemented

| Risk | Mitigation |
|------|------------|
| **AI hallucination** | Model prediction is deterministic (XGBoost); LLM only adds explanation layer |
| **Over-reliance** | App clearly labels AI-generated content; model prediction shown separately |
| **Privacy** | No real employee data sent to API; demo uses synthetic profiles |
| **Bias amplification** | LLM prompted to be "compassionate" and avoid stigmatising language |
| **Confidentiality** | Chat history not persisted; each session is independent |

### Key Design Decision

The GenAI layer is **additive, not core**. The burnout prediction works without the LLM — Claude adds explanations and recommendations on top. If the API is unavailable, the core prediction still functions. This is intentional: the ML model is the source of truth; the LLM is the communication layer.
